```python
import numpy as np
from pyDOE import lhs
import matplotlib.pyplot as plt

def generate_lhs(num_params, ranges, num_samples):
    # Genera i limiti per ciascun parametro
    limits = np.zeros((num_params, 2))
    for i, (min_val, max_val) in enumerate(ranges):
        limits[i] = [min_val, max_val]
    
    # Esegue il campionamento LHS
    samples = lhs(num_params, samples=num_samples, criterion='maximin', iterations=1000)
    
    # Scala i campioni ai limiti specificati
    for i in range(num_params):
        samples[:, i] = samples[:, i] * (limits[i, 1] - limits[i, 0]) + limits[i, 0]
    
    return samples

def plot_lhs_samples(lhs_samples):
    num_params = lhs_samples.shape[1]
    fig, ax = plt.subplots(figsize=(8, 8))
    ax.scatter(lhs_samples[:, 0], lhs_samples[:, 1], color='blue', alpha=0.5)
    ax.set_xlabel('Parametro 1')
    ax.set_ylabel('Parametro 2')
    ax.set_title('Grafico di dispersione per campioni LHS')
    plt.grid(True)
    plt.show()
    
    
def main():
    # Chiedi all'utente il numero di parametri, gli intervalli e il numero di campioni
    num_params = int(input("Inserisci il numero di parametri: "))
    ranges = []
    for i in range(num_params):
        min_val = float(input(f"Inserisci il valore minimo per il parametro {i+1}: "))
        max_val = float(input(f"Inserisci il valore massimo per il parametro {i+1}: "))
        ranges.append((min_val, max_val))
    num_samples = int(input("Inserisci il numero di campioni da generare: "))
    
    # Stampa il numero di campioni da generare
    print(f"\nGenereremo {num_samples} campioni LHS.")
    
    # Genera i campioni LHS
    lhs_samples = generate_lhs(num_params, ranges, num_samples)
    
    # Stampa i campioni generati
    print("\nCampioni generati:")
    for idx, sample in enumerate(lhs_samples, 1):
        print(f"Campioni {idx}: {sample}")
        
    # Genera il plot dei campioni LHS
    plot_lhs_samples(lhs_samples)
        
main()
```

```python
# Definizione della funzione P(x)
def P(x,r,s,t):
        return (r/s) * np.exp(-x/s) +(1-r)*(x/(t**2)) * np.exp(-x/t)

    # Parametri a caso
r = 1.0
s = 0.5
t = 0.8

# Genera valori di x da 0 a 3
x_values = np.linspace(0, 3, 10)

# Calcola i valori corrispondenti di P(x)
p_values = P(x_values, r, s, t)

# Plotta la funzione P(x)
plt.plot(x_values, p_values)
plt.xlabel('x')
plt.ylabel('P(x)')
plt.title('Plot della funzione P(x)')
plt.grid(True)
plt.show()


def generate_lhs_samples(num_samples, ranges):

    lhs_samples = lhs(3, samples=num_samples)  # 3 parametri per la funzione P(x)
    param_samples = []
    for i in range(3):
        min_val, max_val = ranges[i]
        param_values = lhs_samples[:, i] * (max_val - min_val) + min_val
        param_samples.append(param_values)
    return param_samples

# Funzione per chiedere i range per i parametri
def get_param_ranges():
    ranges = []
    for param in ['r', 's', 't']:
        min_val = float(input(f"Inserisci il valore minimo per il parametro {param}: "))
        max_val = float(input(f"Inserisci il valore massimo per il parametro {param}: "))
        ranges.append((min_val, max_val))
    return ranges

# Chiedi all'utente i range per i parametri
print("Inserisci i range per i tre parametri della funzione P(x):")
params_ranges = get_param_ranges()

# Genera campioni LHS per i parametri
num_samples = 10  # Puoi modificare questo valore per ottenere pi o meno campioni
param_samples = generate_lhs_samples(num_samples, params_ranges)

# Genera e plotta le funzioni P(x) corrispondenti ai campioni LHS
plt.figure(figsize=(10, 6))
x_values = np.linspace(0, 3, 10)
for i in range(num_samples):
    p_values = P(x_values, param_samples[0][i], param_samples[1][i], param_samples[2][i])
    plt.plot(x_values, p_values, label=f'Funzione {i+1}')
plt.xlabel('x')
plt.ylabel('P(x)')
plt.title('Plot di funzioni P(x) generate da campioni LHS')
plt.legend()
plt.grid(True)
plt.show()


print("Forma di p_functions:", p_values.shape)
```
