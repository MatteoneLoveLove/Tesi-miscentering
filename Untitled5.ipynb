```python
import numpy as np
import matplotlib.pyplot as plt
import emcee
from scipy.optimize import minimize

# Genera 10000 punti per ogni gaussiana
num_points = 10000

# Genera punti casuali da due distribuzioni normali (gaussiane)
# con media 0 e deviazione standard 1
Gauss_x = np.random.normal(loc=0, scale=1, size=num_points)
Gauss_y = np.random.normal(loc=0, scale=1, size=num_points)

# Associa ad ogni punto della gaussiana un errore dato da una gaussiana di dev 0.5
x = np.random.normal(loc=Gauss_x, scale=0.5, size=num_points)
y = np.random.normal(loc=Gauss_y, scale=0.5, size=num_points)

# Osserviamo la distanza dall'origine
r=np.sqrt(x**2+y**2)



# Plot dei punti gaussiani
plt.figure(figsize=(8, 6))
plt.scatter(x, y, s=5, alpha=0.5)
plt.title('Distribuzione di 10000 punti generati da due gaussiane standard')
plt.xlabel('X')
plt.ylabel('Y')
plt.grid(True)
plt.show()

# Osserva la distribuzione delle distanze
plt.figure(figsize=(8, 6))
plt.hist(r, bins=50, density=True, alpha=0.7)
plt.title('Distribuzione delle distanze dall\'origine')
plt.xlabel('Distanza dall\'origine')
plt.ylabel('Densit')
plt.grid(True)
plt.show()

```

```python
# Definizione della funzione di densit di probabilit per la distribuzione di Rayleigh
def rayleigh_pdf(x, sigma):
    return (x / sigma**2) * np.exp(-x**2 / (2 * sigma**2))

# Logaritmo della funzione di verosimiglianza
def log_likelihood(data, sigma):
    return np.sum(np.log(rayleigh_pdf(data, sigma)))

# Logaritmo del posterior
def log_posterior(sigma, data):
    if sigma <= 0:
        return -np.inf
    return log_likelihood(data, sigma)

# Definizione della funzione negativa del log-likelihood
def NegLL(sigma, data):
    return -log_likelihood(data, sigma)



# Ottimizzazione per trovare il valore iniziale del parametro sigma
result = minimize(NegLL, x0=1.0, args=(r,))
sigma_initial = result.x[0]

# Definizione del numero di walkers e dei passi di catena
ndim = 1
nwalkers = 16
nsteps = 1000

# Posizioni iniziali dei walkers
pos = sigma_initial + 1e-3 * np.random.randn(nwalkers, ndim)

# Avvio dell'algoritmo MCMC
sampler = emcee.EnsembleSampler(nwalkers, ndim, log_posterior, args=[r])
sampler.run_mcmc(pos, nsteps, progress=True)

# Estrazione delle catene di Markov per il parametro sigma
samples = sampler.get_chain(flat=True)


# Calcola la media dei valori di sigma escludendo i primi passi di burn-in per ogni walker
sigma_estimated = np.mean(samples[60,:])

# Plot dell'andamento dei walker
plt.figure(figsize=(10, 6))
for i in range(nwalkers):
    plt.plot(sampler.get_chain()[:, i, 0], label=f"Walker {i + 1}")
plt.xlabel("Step")
plt.ylabel("Sigma")
plt.title("Andamento dei walker durante le MCMC")
plt.legend()
plt.grid(True)
plt.show()


# Plot dell'istogramma delle distanze e della distribuzione di Rayleigh stimata
plt.figure(figsize=(8, 6))
plt.hist(r, bins=50, density=True, alpha=0.7, label='Distanze dall\'origine')
plt.plot(np.linspace(0, 6, 1000), rayleigh_pdf(np.linspace(0, 6, 1000), sigma_estimated), 'r-', label='Distribuzione di Rayleigh stimata')
plt.title('Confronto tra istogramma delle distanze e distribuzione di Rayleigh stimata')
plt.xlabel('Distanza dall\'origine')
plt.ylabel('Densit')
plt.legend()
plt.grid(True)
plt.show()

# Stampiamo il valore stimato di sigma
print("Parametro sigma stimato della distribuzione di Rayleigh:", sigma_estimated)
```
